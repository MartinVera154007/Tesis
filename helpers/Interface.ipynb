## Inicio del c√≥digo de Python

```
from tensorflow.keras.preprocessing import  image
from tensorflow.keras.models import  load_model
import numpy as np
import operator

# Funcion para predecir una imagen, recibe la ruta de la imagen
def predict_image(file):
  img = image.load_img(file, target_size=(256,256))
  img_array = image.img_to_array(img)
  img_array *= 1.0/255.0 
  img_array_expanded_dims = np.expand_dims(img_array, axis=0)
  
  predictions = model.predict(img_array_expanded_dims)
  predictions_list = predictions.tolist()[0]
  index, value = max(enumerate(predictions_list), key=operator.itemgetter(1))
  return classes[index]

# Clases que puede predecir el modelo
classes = ['Complementary Colors', 
           'Duotones',
           'HDR',
           'Image Grain',
           'Light On White',
           'Long Exposure',
           'Macro', 
           'Motion Blur',
           'Negative Image',
           'Rule of thirds',
           'Shallow DOF',
           'Silhouttes',
           'Soft Focus',
           'Vanishing Point'
]

# Cargar el modelo 
model = load_model('/content/drive/MyDrive/Trained Models/ava_aug_full_14_10_epochs_5635.h5')

# Recibir la ruta de la imagen y mandar el estilo en el que se clasifico
while(True):
  try:
    print('Ingrese la ruta de una imagen o "terminar" para salir')
    image_path = input()
    if image_path == 'terminar':
      break
    else:
      estilo = predict_image(image_path)
      print('La imagen es del estilo: ' + estilo)
  except:
    print('Ha ocurrido un error, intentelo de nuevo...')
```
