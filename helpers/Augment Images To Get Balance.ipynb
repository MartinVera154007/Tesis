## Inicio del cÃ³digo de Python

```python
import os
import cv2
import math
from google.colab.patches import cv2_imshow
```

```python
directorio_actual = os.getcwd() # directorio actual de trabajo
directorio_de_imagenes = directorio_actual + '/ava2/content/valid' # directorio donde se encuentran las imagenes
lista_directorios = os.listdir(directorio_de_imagenes) # lista de directorios (clases)
print(lista_directorios)
```

```python
# Definir el porcentaje de imagenes que se van a aumentar por clase
mask = [0, 0, .4084, 0, .05374, .1834, 0, 0, .0427, 0, .4836, 1.5252, .6420, .1904]
i = 0
for directorio in lista_directorios:
  # Si el directorio no se va a aumentar; continuar con el siguiente.
  if mask[i] == 0:
    i += 1
    continue

  directorio_completo = directorio_de_imagenes + '/' + directorio
  imagenes_en_directorio = os.listdir(directorio_completo)
  imagenes_por_aumentar = math.ceil(len(imagenes_en_directorio) * mask[i])
  print(imagenes_por_aumentar)
  if imagenes_por_aumentar == 0: continue
  
  imagenes_aumentadas = 0
  for imagen_nombre in imagenes_en_directorio:
    # Leer la imagen 
    img = cv2.imread(directorio_completo + '/' + imagen_nombre)

    # Espejo horizontal
    horizontal_flipped_image = cv2.flip(img, 1)
    cv2.imwrite(directorio_completo + '/h_' + imagen_nombre, horizontal_flipped_image)
    imagenes_aumentadas += 1
    if imagenes_aumentadas > imagenes_por_aumentar:
      break

    # Espejo vertical
    vertical_flipped_image = cv2.flip(img, 0)
    cv2.imwrite(directorio_completo + '/v_' + imagen_nombre, vertical_flipped_image)
    imagenes_aumentadas += 1
    if imagenes_aumentadas > imagenes_por_aumentar:
      break
  i += 1
```
